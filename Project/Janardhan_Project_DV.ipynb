```python
import pandas as pd
import numpy as np
import matplotlib

# set matplotlib global font size to larger value
matplotlib.rcParams.update({'font.size': 12})

# read dataset into dataframe
pie_df = pd.read_csv("adult.data", header=None, names=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "label"])
pie_df.head()

colums = list(pie_df.columns)
colums.remove("label")
colums.remove("fnlwgt")
stats = [pie_df.groupby([col, "label"])["label"].count().unstack() for col in colums]
#Bar chart for race who are obve 50k and below50k
stats[7].reset_index()
axis = stats[7].reset_index().plot(kind="bar", x="race", color=["deepskyblue", "gold"], stacked=True, figsize=(9,6))
axis.set_xlabel("Race", fontsize=20, fontweight="bold")
axis.set_ylabel("Observations", fontsize=20, fontweight="bold")
axis.get_legend().set_title("Income")
#pie chart on total numnber of etnicity in the study
dataframe = pie_df["race"].value_counts().reset_index()
dataframe.columns = ["race", "frequency"]
axis = dataframe.plot.pie(y="frequency", labels=dataframe.race, colors=["red", "green", "yellow", "skyblue", "grey"], explode=(0.1, 0, 0, 0, 0), shadow=True, autopct='%1.0f%%', startangle=-154, figsize=(12,12))
axis.set_ylabel('')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
import seaborn as sns

dataframe = pd.read_csv('adult.data', header=None)
dataframe[dataframe == '?'] = np.nan
occupationtype=dataframe[[6,14]]
#computation="system"
occupationtype=occupationtype.rename(columns={6: 'Occupation', 14: 'Income'})

fig, axes = plt.subplots(ncols=1, nrows=1, figsize=(15,7))
mosaic(occupationtype, ['Occupation', 'Income'], ax=axes, axes_label=False)
plt.title('Occupation VS Salary', fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from datetime import timedelta


dataframe = pd.read_csv(
    r"adult.data",
    index_col=None,
    header=None,
)
dataframe.columns = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education-num",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
    "native-country",
    "income",
]
dataframe["income"] = dataframe["income"].apply(lambda x: 1 if x == " >50K" else 0)
dataframe["Wife"] = dataframe["relationship"].apply(lambda x: 1 if x == " Wife" else 0)
dataframe["Own-child"] = dataframe["relationship"].apply(lambda x: 1 if x == " Own-child" else 0)
dataframe["Husband"] = dataframe["relationship"].apply(lambda x: 1 if x == " Husband" else 0)
dataframe["Not-in-family"] = dataframe["relationship"].apply(
    lambda x: 1 if x == " Not-in-family" else 0
)
dataframe["Other-relative"] = dataframe["relationship"].apply(
    lambda x: 1 if x == " Other-relative" else 0
)
dataframe["Unmarried"] = dataframe["relationship"].apply(lambda x: 1 if x == " Unmarried" else 0)
# relationship analysis
performance = []

wife_df = dataframe[dataframe["Wife"] == 1]
wife_income_df = wife_df[wife_df["income"] == 1]
# print(len(wife_income_df) / len(wife_df))
performance.append(len(wife_income_df) / len(wife_df))

own_child_df = dataframe[dataframe["Own-child"] == 1]
own_child_income_df = own_child_df[own_child_df["income"] == 1]
# print(len(own_child_income_df) / len(own_child_df))
performance.append(len(own_child_income_df) / len(own_child_df))
not_in_family_df = dataframe[dataframe["Not-in-family"] == 1]
not_in_family_income_df = not_in_family_df[not_in_family_df["income"] == 1]
# print(len(not_in_family_income_df) / len(not_in_family_df))
performance.append(len(not_in_family_income_df) / len(not_in_family_df))

Husband_df = dataframe[dataframe["Husband"] == 1]
Husband_income_df = Husband_df[Husband_df["income"] == 1]
# print(len(Husband_income_df) / len(Husband_df))
performance.append(len(Husband_income_df) / len(Husband_df))

unmarried_df = dataframe[dataframe["Unmarried"] == 1]
unmarried_income_df = unmarried_df[unmarried_df["income"] == 1]
# print(len(unmarried_income_df) / len(unmarried_df))
performance.append(len(unmarried_income_df) / len(unmarried_df))

other_relative_df = dataframe[dataframe["Other-relative"] == 1]
other_relative_income_df = other_relative_df[other_relative_df["income"] == 1]
# print(len(other_relative_income_df) / len(other_relative_df))
performance.append(len(other_relative_income_df) / len(other_relative_df))



income_df = dataframe[dataframe["income"] == 1]
# print(len(income_df) / len(df))
average = len(income_df) / len(dataframe)

labels = (
    "Wife",
    "Own-child",
        "Not-in-family",
    "Husband",
 "Unmarried",
    "Other-relative",
   
)
y_pos = np.arange(len(labels))

plt.figure(figsize=(8, 6))
plt.bar(y_pos, performance, align="center", alpha=0.5)
plt.xticks(y_pos, labels)
plt.ylabel("% >50k")
for i in range(len(labels)):
    plt.text(i, performance[i], round(performance[i], 2), horizontalalignment="center")
plt.title("Income vs. Relationship Status")
plt.grid()
average_label = "average (" + str(round(average, 4) * 100) + "%)"
plt.axhline(y=average, linewidth=2, color="r", label=average_label)
plt.legend(loc="upper right")
plt.savefig("relationship.png")
```
